```python
import csv
import requests
from bs4 import BeautifulSoup

def search_flipkart(keyword):
    next_page=True
    Page=1
    f = open('flipkartWebScraperDataset.csv', 'a+', newline='', encoding='UTF8')
    writer = csv.writer(f)
    while next_page:
        # Set the flipkart URL with the search keyword
        url = f'https://www.flipkart.com/search?q={keyword.replace(" ", "+")}&page={Page}'

        # Send a GET request to the URL
        response = requests.get(url)

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the product listings on the page 
        nav_tag = soup.find('nav', {'class': 'yFHi8N'})
        product_name = soup.find_all('div',{'class':'_4rR01T'})
        product_price = soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
        product_rating = soup.find_all('div',{'class':'_3LWZlK'})

        # Iterate over the product listings and extract relevant information
        for items in range(0,len(product_name)):
            # Extract the product title
            title = product_name[items].text.strip()

            # Extract the product price
            price = product_price[items].text.strip()

            # Extract the product rating
            rating = product_rating[items].text.strip()

            header = ['Title','Price','Rating']
            data = [title,price,rating]
            # Check if the file is empty (if it doesn't contain any rows)
            if f.tell() == 0:
                writer.writerow(header)  # Write the header row

            writer.writerow(data)  # Write the data row

            # Print the product details
#             print('Title:', title)
#             print('Price:', price)
#             print('Rating:', rating)
#             print('---')
        Page+=1
        # if next_page exists, it will update the url and continue to execute the above steps
        if product_name:
            continue
#         if (nav_tag.find('a',{'class':'_1LKTO3'}).text=='Next'):
#             print('about to enter next_page')
#             continue
        else:
            next_page=False
    f.close()


# Call the search_flipkart function with the keyword 'vivo mobile'
search_flipkart('vivo mobile')
```
